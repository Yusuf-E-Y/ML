### Cuda , Float32

```python
matrix = torch.tensor([[1,2,3],[4,5,6]],device='cuda',dtype=torch.float32)
matrix.device,matrix.dtype,matrix.ndim
```

### Cuda , uint8

```python
matrix = torch.randint(0,255,(256,256,3), dtype=torch.uint8, device='cuda') #Burada 256x256 resim düşünürsek bunu cudaya taşımış oluruz
matrix.device
```

### Torch --> Numpy

```python
np_matrix = matrix.cpu() #Cuda --> CPU
np_matrix = np_matrix.numpy() #numpy arrayine çevrilir
np_matrix.shape, np_matrix.dtype    
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.imshow(np_matrix, cmap='gray')
plt.axis('off')
plt.show()  
```
